**Explore the relationship between Car Weight and CO2 emissions**

```python
# scatterplot of car weight and co2 emissions
bar_plot = sns.lmplot(x='Weight', y='CO2', data=df, height=10)

ax = plt.gca()

ax.set_title("Relationship Between Car Weight and CO2 emissions")

# Set label for x-axis
ax.set_xlabel( "Car Weight" , size = 12 )
  
# Set label for y-axis
ax.set_ylabel( "CO2 Emissions" , size = 12 )

plt.show();

```

- Scatterplot suggests a strong positive relationship between Car weight and Co2 emissions

**Next we will explore the relationship between car volume and the CO2 emssions**

```python
# scatterplot of car volume and co2 emissions
bar_plot = sns.lmplot(x='Volume', y='CO2', data=df, height=10)

ax = plt.gca()

ax.set_title("Relationship Between Car Volume and CO2 emissions")

# Set label for x-axis
ax.set_xlabel( "Car Volume" , size = 12 )
  
# Set label for y-axis
ax.set_ylabel( "CO2 Emissions" , size = 12 )

plt.show();

```

- Scatterplot suggests a strong positive relationship between Car Volume and Co2 emissions

### Null and Alternative Hypotheses

**Null Hypothesis:** There is no connection between car weight, volume, and CO2 emissions. It suggests that the weight and volume of a car do not have any effect on the amount of CO2 emissions it produces.

**Alternative Hypothesis:** There is a relationship between car weight, volume, and CO2 emissions. It proposes that the weight and volume of a car do have an impact on the amount of CO2 emissions it produces.

The Type I error rate will be set at 5% which means that if the p-value is more than 0.05 then we would fail to reject the null hypothesis. On the other hand if the p-value is less than 0.05 then would reject the null hypothesis and accept the alternative hypothesis. 

```python
# Create A Multiple Linear Regression Model
import statsmodels.api as sm

model = sm.OLS(df['CO2'], df[['Weight', 'Volume']])
results = model.fit()
results.summary()
```
